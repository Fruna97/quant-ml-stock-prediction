```python
'''
웹사이트의 테이블을 데이터프레임으로 읽기
'''
import pandas as pd
import re


# Wikipedia에서 S&P 500 테이블 가져오기
# url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
# url = "https://en.wikipedia.org/wiki/List_of_S%26P_400_companies"
url = "https://en.wikipedia.org/wiki/List_of_S%26P_600_companies"
data = pd.read_html(url)
data = data[0]

# 컬럼명 소문자로 변경
data.columns = [re.sub(r'[^\w]', '_', col) for col in data.columns.str.lower()]

# data['date_added'] = pd.to_datetime(data['date_added'])

display(data)
```

```python
'''
CSV 파일로 저장
'''
data.to_csv('Data/Stock/S&P600.csv', index=False)
```

```python
'''
DB에 S&P500 종목들 삽입
'''
import sqlite3
import pandas as pd


conn = sqlite3.connect('Database/S&P500.db')

data.to_sql('stock', conn, if_exists='replace', index=False)

conn.close()
```

```python
'''
DB의 데이터 확인
'''
import sqlite3
import pandas as pd


conn = sqlite3.connect('Database/S&P500.db')
query = "SELECT * FROM stock;"
stock = pd.read_sql_query(query, conn)
conn.close()
display(stock)
```
