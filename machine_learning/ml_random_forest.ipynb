```python
'''
모델링 및 검증 : RandomForest, k-fold 사용
'''
import pandas as pd
import numpy as np
import time
from sklearn.model_selection import KFold
from cuml import RandomForestClassifier
from cuml import RandomForestRegressor
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score


for sector, data in dps.items():    
    # data = data[['date', 'symbol', 'gics_sector', 'volume', 'sma_5', 'sma_10', 'sma_20', 'ema_5', 'ema_10', 'ema_20', 'change']]
    data = data[['date', 'symbol', 'gics_sector', 'volume', 'sma_5', 'sma_10', 'sma_20', 'change']]


    # 업종 컬럼 삭제
    data = data.drop(columns=['gics_sector'])


    # 종목 컬럼 삭제 or 종목이름 원핫인코딩
    data = data.drop(columns=['symbol'])


    # 학습, 검증 데이터 분리
    # x_train = data.loc[data['date'].between('2001-01-01', '2022-01-01', inclusive='left') | data['date'].between('2023-01-01', '2034-01-01', inclusive='left')].iloc[ : ,  :-1]
    # x_test = data.loc[data['date'].between('2022-01-01', '2023-01-01', inclusive='left')].iloc[ : ,  :-1]

    # y_train_r = data.loc[data['date'].between('2001-01-01', '2022-01-01', inclusive='left') | data['date'].between('2023-01-01', '2034-01-01', inclusive='left')].iloc[ : , -1]
    # y_test_r = data.loc[data['date'].between('2022-01-01', '2023-01-01', inclusive='left')].iloc[ : , -1]
    x = data.iloc[ : ,  :-1]
    y = data.iloc[ : , -1]
    
    
    # 날짜 컬럼 삭제
    x = x.drop(columns=['date'])
    
    
    # cuml의 waring에 의한 설정
    x = x.astype('float32')
    y = y.astype('float32')

    
    # 모델 학습
    # k-fold 설정
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    accuracy_scores = []
    precision_scores = []
    recall_scores = []
    f1_scores = []
    for train_index, test_index in kf.split(x):
        x_train, x_test = x.iloc[train_index], x.iloc[test_index]
        y_train_r, y_test_r = y.iloc[train_index], y.iloc[test_index]
        y_train_c = y_train_r > 0
        y_test_c = y_test_r > 0
    
        model_r = RandomForestRegressor(
            n_estimators=750, # default : 100
            split_criterion='mse', # default : 'mse', node 나눌 때 어떤 격차 계산 기반으로 나눌지
            bootstrap=True, # default : True, 각 숲이 부트스프랩된 샘플 기반으로 만들어질지 여부, False면 각 숲이 전체 데이터 기반으로 만들어짐
            max_samples=1.0, # default : 1.0, 각 트리를 적합하는 데 사용된 데이터셋 행의 비율
            max_depth=64, # default : 16, 각 숲당 최대깊이
            max_leaves=-1, # default : -1, 트리당 최대 리프 노드
            max_features=1.0, # default : 1.0, 노드를 분할 할 때 사용할 열의 수. float이면 비율로써 사용됨
            n_bins=256, # default : 128, 분할 알고리즘에서 사용하는 최대 빈 수. 비대칭일 경우 늘리면 정확도가 향상 될 수 있음.
            n_streams=100, # default : 4, 숲 조성에 사용되는 병렬 스트림 수
            min_samples_leaf=1, # default : 1, 각 리프 노드가 가지는 최소 샘플 수
            min_samples_split=10, # default : 2, 노드를 분할 하는 데 필요한 최소 샘플 수
            min_impurity_decrease =0.0, # default : 0.0, 분할 할 때 필요한 최소 불순물 감소량
            # accuracy_metric ='r2', # default : 'r2', 모델 성능 평가할 때 사용되는 지표
            max_batch_size=4096, # default : 4096, 주어진 배치에서 처리할 수 있는 최대 노드 수
        )
        model_r.fit(x_train, y_train_r)
        pred_r = model_r.predict(x_test)
        pred_c = pd.Series(pred_r) > 0
    
        # 점수화
        acs = accuracy_score(y_test_c, pred_c)
        prs = precision_score(y_test_c, pred_c)
        res = recall_score(y_test_c, pred_c)
        f1s = f1_score(y_test_c, pred_c)
        accuracy_scores.append(acs)
        precision_scores.append(prs)
        recall_scores.append(res)
        f1_scores.append(f1s)
    print('SECTOR : ', sector)
    print('accuracy_score : ', np.mean(accuracy_scores))
    print('precision_score : ', np.mean(precision_scores))
    print('recall_score : ', np.mean(recall_scores))
    print('f1_score : ', np.mean(f1_scores))
```
