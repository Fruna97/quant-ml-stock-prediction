```python
import pandas as pd
from bokeh.plotting import figure
from bokeh.models import Range1d, LinearAxis, DataRange1d
from bokeh.palettes import Category10
import itertools

def create_wheel_zoomable_plot(df, y_columns, width=800, height=600, title="Wheel Zoomable Chart"):
    _df = df
    
    # Bokeh 그래프 생성
    p = figure(
        width=width,
        height=height,
        title=title,
        # x_axis_type="datetime", 
        tools="xwheel_zoom,pan,reset",
        active_scroll="xwheel_zoom",
    )
    
    p.extra_y_ranges = {
        # 'Close' : DataRange1d(), 
        'autorange1' : DataRange1d(),
        'indicator1' : Range1d(-20, 20),
        'indicator2' : Range1d(0, 1), 
        'indicator3' : Range1d(0, 100),
        'indicator4' : Range1d(-100, 0),
        'indicator5' : Range1d(-1, 1),
        'indicator6' : Range1d(-100, 100),
        'indicator7' : Range1d(-50, 50),
        'call' : Range1d(-1, 1)
    }
    # p.add_layout(LinearAxis(y_range_name='Close'), 'left')
    p.add_layout(LinearAxis(y_range_name='autorange1'), 'right')
    p.add_layout(LinearAxis(y_range_name='indicator1'), 'right')
    p.add_layout(LinearAxis(y_range_name='indicator2'), 'right')
    p.add_layout(LinearAxis(y_range_name='indicator3'), 'right')
    p.add_layout(LinearAxis(y_range_name='indicator4'), 'right')
    p.add_layout(LinearAxis(y_range_name='indicator5'), 'right')
    p.add_layout(LinearAxis(y_range_name='indicator6'), 'right')
    p.add_layout(LinearAxis(y_range_name='indicator7'), 'right')
    p.add_layout(LinearAxis(y_range_name='call'), 'right')
    
    # 각 컬럼에 대해 라인 그래프 추가
    colors = itertools.cycle(Category10[10])
    for col in y_columns:
        if col == 'Close' or col == 'kama' or col == 'kama_signal' or col == 'ema_9' or col == 'vwap' or col == 'vwap_signal' or col == 'bollinger_hband' or col == 'bollinger_lband' or col == 'donchian_channel_hband' or col == 'donchian_channel_lband' or col == 'keltner_channel_hband' or col == 'keltner_channel_lband' or col == 'ichimoku_a' or col == 'ichimoku_b' or col == 'ichimoku_conversion_line' or col == 'ichimoku_base_line' or col == 'psar' or col == 'psar_ema' or col == 'psar_smoothing' or col == 'pp1' or col == 'pp2':
            r = p.line(_df.reset_index().index, _df[col], legend_label=col, color=next(colors))
            p.y_range.renderers = [r]
            # r = p.line(_df.reset_index().index, _df[col], legend_label=col, color=next(colors), y_range_name='Close')
            # p.extra_y_ranges['Close'].renderers = [r]
        elif col.find('_call') != -1:
            r = p.line(_df.reset_index().index, _df[col], legend_label=col, color=next(colors), y_range_name='call')
        elif col == 'adi' or col == 'adi_signal' or col == 'fi' or col == 'fi_signal' or col == 'nvi' or col == 'nvi_signal_short' or col == 'nvi_signal_long' or col == 'obv_signal' or col == 'obv' or col == 'vpt_signal' or col == 'vpt' or col == 'bollinger_wband' or col == 'donchian_channel_wband' or col == 'keltner_channel_wband':
            r = p.line(_df.reset_index().index, _df[col], legend_label=col, color=next(colors), y_range_name='autorange1')
        elif col == 'adx' or col == 'mi' or col == 'call':
            r = p.line(_df.reset_index().index, _df[col], legend_label=col, color=next(colors), y_range_name='indicator7')
        elif col == 'eom' or col == 'eom_sma' or col == 'aroon_up' or col == 'aroon_down' or col == 'cci' or col == 'cci_signal' or col == 'kst' or col == 'kst_signal':
            r = p.line(_df.reset_index().index, _df[col], legend_label=col, color=next(colors), y_range_name='indicator6')
        elif col == 'cmf' or col == 'cmf_signal' or col == 'bollinger_pband' or col == 'bollinger_pband_signal' or col == '0.8' or col == '0.2' or col == 'donchian_channel_pband' or col == 'donchian_channel_pband_signal' or col == 'keltner_channel_pband' or col == 'keltner_channel_pband_signal' or col == 'adx_pos' or col == 'adx_neg' or col == 'trix' or col == 'trix_signal' or col == 'vi_pos' or col == 'vi_neg':
            r = p.line(_df.reset_index().index, _df[col], legend_label=col, color=next(colors), y_range_name='indicator5')
        elif col == 'williams_r' or col == 'williams_r_signal' or col == '-80' or col == '-20':
            r = p.line(_df.reset_index().index, _df[col], legend_label=col, color=next(colors), y_range_name='indicator4')
        elif col == 'uo' or col == 'uo_signal' or col == 'stoch_k' or col == 'stoch_d' or col == '30' or col == '70' or col == 'mfi' or col == 'mfi_signal' or col == 'stc':
            r = p.line(_df.reset_index().index, _df[col], legend_label=col, color=next(colors), y_range_name='indicator3')
        elif col == 'stochrsi' or col == 'stochrsi_signal' or col == 'stochrsi_fast_k' or col == 'stochrsi_slow_k' or col == 'stochrsi_d':
            r = p.line(_df.reset_index().index, _df[col], legend_label=col, color=next(colors), y_range_name='indicator2')
        else:
            r = p.line(_df.reset_index().index, _df[col], legend_label=col, color=next(colors), y_range_name='indicator1')
            
    # x축 설정
    # p.xaxis.formatter.days = "%Y-%m-%d %H:%M:%S" # x축 레이블 포맷 설정
    # p.xaxis.major_label_orientation = 0.5  # x축 레이블 기울기 설정
    p.xaxis.major_label_overrides = {
        i: date.strftime('%Y-%m-%d %H:%S') for i, date in enumerate(_df.index)
    }
    
    p.x_range.range_padding = 0.1
    p.y_range.range_padding = 0.1
    p.x_range.range_padding_units = 'percent'
    p.y_range.range_padding_units = 'percent'
    
    p.legend.location = "top_left"
    p.legend.click_policy = "hide"
    
    return p
```

```python
import warnings
import importlib
import pandas as pd
import ta.momentum
import ta.volume
import ta.volatility
import ta.trend
import Custom.CustomizedAnalysis as ca
import matplotlib.pyplot as plt
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, reset_output


warnings.filterwarnings('ignore', category=pd.errors.PerformanceWarning)
warnings.filterwarnings('ignore', category=FutureWarning)
importlib.reload(ca)


for sector, tickers in ticker_by_sector.items():
    for ticker in tickers:
        # 데이터 불러오기
        try:
            data = pd.read_csv(f'Data/Price/Fix/{ticker}_price_5min.csv', header=[0, 1], index_col=0)
        except FileNotFoundError:
            continue
        
        
        # 데이터 정리
        data = data.xs(ticker, level=1, axis=1)
        data.index = pd.to_datetime(data.index)
        data = data.sort_index(ascending=True)
        
        
        ''' 이동평균 (Moving Average) '''
        # # 단순이동평균 SMA
        # data['sma_5'] = ta.trend.sma_indicator(data['Close'], window=5 * 78)
        # data['sma_10'] = ta.trend.sma_indicator(data['Close'], window=10 * 78)
        # data['sma_20'] = ta.trend.sma_indicator(data['Close'], window=20 * 78)
        # data['sma_50'] = ta.trend.sma_indicator(data['Close'], window=50 * 78)
        # data['sma_60'] = ta.trend.sma_indicator(data['Close'], window=60 * 78)
        # data['sma_120'] = ta.trend.sma_indicator(data['Close'], window=120 * 78)
        # data['sma_200'] = ta.trend.sma_indicator(data['Close'], window=200 * 78)

        # # 가중이동평균 WMA
        # data['wma_5'] = ta.trend.wma_indicator(data['Close'], window=5 * 78)
        # data['wma_10'] = ta.trend.wma_indicator(data['Close'], window=10 * 78)
        # data['wma_20'] = ta.trend.wma_indicator(data['Close'], window=20 * 78)
        # data['wma_50'] = ta.trend.wma_indicator(data['Close'], window=50 * 78)
        # data['wma_60'] = ta.trend.wma_indicator(data['Close'], window=60 * 78)
        # data['wma_120'] = ta.trend.wma_indicator(data['Close'], window=120 * 78)
        # data['wma_200'] = ta.trend.wma_indicator(data['Close'], window=200 * 78)

        # # 지수이동평균 EMA
        data['ema_5'] = ta.trend.ema_indicator(data['Close'], window=5 * 78)
        # data['ema_10'] = ta.trend.ema_indicator(data['Close'], window=10 * 78)
        # data['ema_20'] = ta.trend.ema_indicator(data['Close'], window=20 * 78)
        # data['ema_50'] = ta.trend.ema_indicator(data['Close'], window=50 * 78)
        # data['ema_60'] = ta.trend.ema_indicator(data['Close'], window=60 * 78)
        # data['ema_120'] = ta.trend.ema_indicator(data['Close'], window=120 * 78)
        # data['ema_200'] = ta.trend.ema_indicator(data['Close'], window=200 * 78)
        
        
        ''' 모멘텀 지표 (Momentum Indicator) '''
        # Awesome Oscillator
        data['ao'] = ta.momentum.awesome_oscillator(high=data['High'], low=data['Low'], window1=5 * 78, window2=34 * 78)
        data['ao_signal'] = ta.trend.ema_indicator(close=data['ao'], window = 9 * 78)
        data['ao_shift'] = data['ao'].shift(1)
        data['ao_ao_signal_diff'] = data['ao'] - data['ao_signal'] # GC
        data['ao_ao_signal_diff_shift'] = data['ao_ao_signal_diff'].shift(1)
        data['ao_zero_call'] = data.apply(ca.ao_call, axis=1) # GC
        data['ao_signal_call'] = data.apply(ca.ao_signal_call, axis=1) # GC
        data.pop('ao_shift')
        data.pop('ao_ao_signal_diff')
        data.pop('ao_ao_signal_diff_shift')
        
        # KAMA (Kaufman’s Adaptive Moving Average)
        # KAMA의 신호선과의 교차도 사용하자. 이유 : 웹서핑을 해보면 상승/하강할 때 롱/숏 포지션을 가지라고 하는데, 이것은 KAMA의 이동평균선과 KAMA와 비교를 통해서 상승/하강의 여부를 알 수 있기 때문
        data['kama'] = ta.momentum.kama(close=data['Close'], window=10 * 78, pow1=2, pow2=30)
        data['kama_signal'] = ta.trend.ema_indicator(close=data['kama'], window=9 * 78)
        data['kama_close_kama_diff'] = data['Close'] - data['kama'] # GC
        data['kama_close_kama_diff_shift'] = data['kama_close_kama_diff'].shift(1)
        # data['kama_kama_ema_diff'] = data['kama'] - data['ema_9'] # GC
        # data['kama_kama_ema_diff_shift'] = data['kama_kama_ema_diff'].shift(1)
        data['kama_kama_signal_diff'] = data['kama'] - data['kama_signal'] # GC
        data['kama_kama_signal_diff_shift'] = data['kama_kama_signal_diff'].shift(1)
        data['kama_call'] = data.apply(ca.kama_call, axis=1) # GC
        # data['kama_ema_call'] = data.apply(ca.kama_ema_call, axis=1) # GC
        data['kama_signal_call'] = data.apply(ca.kama_signal_call, axis=1) # GC
        data.pop('kama_close_kama_diff')
        data.pop('kama_close_kama_diff_shift')
        # data.pop('kama_kama_ema_diff')
        # data.pop('kama_kama_ema_diff_shift')
        data.pop('kama_kama_signal_diff')
        data.pop('kama_kama_signal_diff_shift')
        
        # PPO (Percentage Price Oscillator)
        ppo = ta.momentum.PercentagePriceOscillator(close=data['Close'], window_slow=26 * 78, window_fast=12 * 78, window_sign=9 * 78)
        data['ppo'] = ppo.ppo()
        data['ppo_signal'] = ppo.ppo_signal()
        data['ppo_hist'] = ppo.ppo_hist() # GC
        data['ppo_hist_shift'] = data['ppo_hist'].shift(1)
        data['ppo_signal_call'] = data.apply(ca.ppo_signal_call, axis=1) # GC
        data.pop('ppo_hist')
        data.pop('ppo_hist_shift')
                
        # PVO (Percentage Volume Oscillator)
        pvo = ta.momentum.PercentageVolumeOscillator(volume=data['Volume'], window_slow=26 * 78, window_fast=12 * 78, window_sign=9 * 78)
        data['pvo'] = pvo.pvo()
        data['pvo_signal'] = pvo.pvo_signal()
        data['pvo_hist'] = pvo.pvo_hist() # GC
        data['pvo_hist_shift'] = data['pvo_hist'].shift(1)
        data['pvo_signal_call'] = data.apply(ca.pvo_signal_call, axis=1) # GC
        data.pop('pvo_hist')
        data.pop('pvo_hist_shift')
        
        # ROC (Rate of Change)
        data['roc'] = ta.momentum.roc(close=data['Close'], window=12 * 78)
        
        # RSI (Relative Strength Index)
        # XXX : RSI를 5분 단위로 구하면 현재 상태가 원래 목적의 숫자만큼 올라가지 않는 듯하다.(ta의 RSI계산에 사용하는 EMA가 원인이라고 추정됨)
        data['rsi'] = ta.momentum.rsi(close=data['Close'], window=14 * 78)
        
        # Stochastic RSI Oscillator
        stoch_rsi = ta.momentum.StochRSIIndicator(close=data['Close'], window=14 * 78, smooth1=3 * 78, smooth2=3 * 78)
        data['stochrsi_fast_k'] = stoch_rsi.stochrsi()
        data['stochrsi_slow_k'] = stoch_rsi.stochrsi_k()
        data['stochrsi_d'] = stoch_rsi.stochrsi_d()
        data['stochrsi_fast_k_shift'] = data['stochrsi_fast_k'].shift(1)
        data['stochrsi_fast_d_diff'] = data['stochrsi_fast_k'] - data['stochrsi_d'] # GC
        data['stochrsi_fast_d_diff_shift'] = data['stochrsi_fast_d_diff'].shift(1)
        data['stochrsi_call'] = data.apply(ca.stochrsi_call, axis=1) # GC
        data['stochrsi_fast_d_call'] = data.apply(ca.stochrsi_fast_d_call, axis=1) # GC
        data.pop('stochrsi_fast_k_shift')
        data.pop('stochrsi_fast_d_diff')
        data.pop('stochrsi_fast_d_diff_shift')
        
        # Stochastic Oscillator
        stoch = ta.momentum.StochasticOscillator(close=data['Close'], high=data['High'], low=data['Low'], window=14 * 78, smooth_window=3 * 78)
        data['stoch_k'] = stoch.stoch()
        data['stoch_d'] = stoch.stoch_signal()
        data['stoch_k_shift'] = data['stoch_k'].shift(1)
        data['stoch_k_d_diff'] = data['stoch_k'] - data['stoch_d'] # GC
        data['stoch_k_d_diff_shift'] = data['stoch_k_d_diff'].shift(1)
        data['stoch_call'] = data.apply(ca.stoch_call, axis=1) # GC
        data['stoch_k_d_call'] = data.apply(ca.stoch_k_d_call, axis=1) # GC
        data.pop('stoch_k_shift')
        data.pop('stoch_k_d_diff')
        data.pop('stoch_k_d_diff_shift')

        # TSI (True Strength Index)
        # -25, 25선 교차와 0선 교차가 있는데, 여기서는 0선 교차를 사용
        data['tsi'] = ta.momentum.tsi(close=data['Close'], window_slow=25 * 78, window_fast=13 * 78)
        data['tsi_shift'] = data['tsi'].shift(1)
        data['tsi_signal'] = ta.trend.ema_indicator(close=data['tsi'], window=9 * 78) # GC
        data['tsi_tsi_signal_diff'] = data['tsi'] - data['tsi_signal'] # GC
        data['tsi_tsi_signal_diff_shift'] = data['tsi_tsi_signal_diff'].shift(1)
        data['tsi_call'] = data.apply(ca.tsi_call, axis=1) # GC
        data['tsi_signal_call'] = data.apply(ca.tsi_signal_call, axis=1) # GC
        data.pop('tsi_shift')
        data.pop('tsi_tsi_signal_diff')
        data.pop('tsi_tsi_signal_diff_shift')
        
        # Ultimate Oscillator
        data['uo'] = ta.momentum.ultimate_oscillator(high=data['High'], low=data['Low'], close=data['Close'], window1=7 * 78, window2=14 * 78, window3=28 * 78)
        data['uo_signal'] = ta.trend.ema_indicator(close=data['uo'], window=9 * 78) # GC
        data['uo_shift'] = data['uo'].shift(1)
        data['uo_uo_signal_diff'] = data['uo'] - data['uo_signal'] # GC
        data['uo_uo_signal_diff_shift'] = data['uo_uo_signal_diff'].shift(1)
        data['uo_call'] = data.apply(ca.uo_call, axis=1) # GC
        data['uo_signal_call'] = data.apply(ca.uo_signal_call, axis=1) # GC
        data.pop('uo_shift')
        data.pop('uo_uo_signal_diff')
        data.pop('uo_uo_signal_diff_shift')
        
        # Williams %R
        data['williams_r'] = ta.momentum.williams_r(high=data['High'], low=data['Low'], close=data['Close'], lbp=14 * 78)
        data['williams_r_signal'] = ta.trend.ema_indicator(close=data['williams_r'], window=9 * 78) #GC
        data['williams_r_shift'] = data['williams_r'].shift(1)
        data['williams_r_r_signal_diff'] = data['williams_r'] - data['williams_r_signal']
        data['williams_r_r_signal_diff_shift'] = data['williams_r_r_signal_diff'].shift(1)
        data['williams_r_call'] = data.apply(ca.williams_r_call, axis=1) # GC
        data['williams_r_signal_call'] = data.apply(ca.williams_r_signal_call, axis=1) # GC
        data.pop('williams_r_shift')
        data.pop('williams_r_r_signal_diff')
        data.pop('williams_r_r_signal_diff_shift')
                

        ''' 거래량 지표 (Volume Indicator) '''
        # ADI (Accumulation/Distribution Index)
        data['adi'] = ta.volume.acc_dist_index(high=data['High'], low=data['Low'], close=data['Close'], volume=data['Volume'])
        data['adi_signal'] = ta.trend.ema_indicator(close=data['adi'], window=9 * 78) #GC
        data['adi_adi_signal_diff'] = data['adi'] - data['adi_signal'] # GC
        data['adi_adi_signal_diff_shift'] = data['adi_adi_signal_diff'].shift(1)
        data['adi_signal_call'] = data.apply(ca.adi_signal_call, axis=1) # GC
        data.pop('adi_adi_signal_diff')
        data.pop('adi_adi_signal_diff_shift')
        
        # CMF (Chaikin Money Flow)
        data['cmf'] = ta.volume.chaikin_money_flow(high=data['High'], low=data['Low'], close=data['Close'], volume=data['Volume'], window=20 * 78)
        data['cmf_signal'] = ta.trend.ema_indicator(close=data['cmf'], window=9 * 78) # GC
        data['cmf_shift'] = data['cmf'].shift(1)
        data['cmf_cmf_signal_diff'] = data['cmf'] - data['cmf_signal'] # GC
        data['cmf_cmf_signal_diff_shift'] = data['cmf_cmf_signal_diff'].shift(1)
        data['cmf_call'] = data.apply(ca.cmf_call, axis=1) # GC
        data['cmf_signal_call'] = data.apply(ca.cmf_signal_call, axis=1) # GC
        data.pop('cmf_shift')
        data.pop('cmf_cmf_signal_diff')
        data.pop('cmf_cmf_signal_diff_shift')
            
        # EoM, EMV (Ease of movement)
        # MEMO : 중간에 Volume이 0이면 결측치가 발생하여 SMA가 14일동안 끊김, 일단 사용
        ease_of_movement = ta.volume.EaseOfMovementIndicator(high=data['High'], low=data['Low'], volume=data['Volume'], window=14 * 78)
        data['eom'] = ease_of_movement.ease_of_movement()
        data['eom_sma'] = ease_of_movement.sma_ease_of_movement()
        data['eom_sma_shift'] = data['eom_sma'].shift(1)
        data['eom_call'] = data.apply(ca.eom_sma_indicator, axis=1)
        data.pop('eom_sma_shift')
        
        # FI (Force Index)
        data['fi'] = ta.volume.force_index(close=data['Close'], volume=data['Volume'], window=13 * 78)
        data['fi_signal'] = ta.trend.ema_indicator(close=data['fi'], window=9 * 78) # GC
        data['fi_fi_signal_diff'] = data['fi'] - data['fi_signal'] # GC
        data['fi_fi_signal_diff_shift'] = data['fi_fi_signal_diff'].shift(1)
        data['fi_signal_call'] = data.apply(ca.fi_signal_call, axis=1)
        data.pop('fi_fi_signal_diff')
        data.pop('fi_fi_signal_diff_shift')
        
        # MFI (Money Flow Index)
        # XXX : RSI와 마찬가지로 5분 단위 데이터로 넘어가면 값이 쪼그라듬, signal call만 사용하자
        data['mfi'] = ta.volume.money_flow_index(high=data['High'], low=data['Low'], close=data['Close'], volume=data['Volume'], window=14 * 78)
        data['mfi_signal'] = ta.trend.ema_indicator(close=data['mfi'], window=9 * 78) # GC
        data['mfi_shift'] = data['mfi'].shift(1)
        data['mfi_mfi_signal_diff'] = data['mfi'] - data['mfi_signal']
        data['mfi_mfi_signal_diff_shift'] = data['mfi_mfi_signal_diff'].shift(1)
        # data['mfi_call'] = data.apply(ca.mfi_call, axis=1)
        data['mfi_signal_call'] = data.apply(ca.mfi_signal_call, axis=1)
        data.pop('mfi_shift')
        data.pop('mfi_mfi_signal_diff')
        data.pop('mfi_mfi_signal_diff_shift')
        
        # NVI (Negative Volume Index)
        data['nvi'] = ta.volume.negative_volume_index(close=data['Close'], volume=data['Volume'])
        data['nvi_signal_short'] = ta.trend.ema_indicator(close=data['nvi'], window=3 * 78) # GC
        data['nvi_signal_long'] = ta.trend.ema_indicator(close=data['nvi'], window=255 * 78) # GC
        data['nvi_short_long_diff'] = data['nvi_signal_short'] - data['nvi_signal_long'] # GC
        data['nvi_short_long_diff_shift'] = data['nvi_short_long_diff'].shift(1)
        data['nvi_short_long_call'] = data.apply(ca.nvi_short_long_call, axis=1) # GC
        data.pop('nvi_short_long_diff')
        data.pop('nvi_short_long_diff_shift')
    
        # OBV (On-Balance Volume)
        data['obv'] = ta.volume.on_balance_volume(close=data['Close'], volume=data['Volume'])
        data['obv_signal'] = ta.trend.ema_indicator(close=data['obv'], window=9 * 78)
        data['obv_obv_signal_diff'] = data['obv'] - data['obv_signal'] # GC
        data['obv_obv_signal_diff_shift'] = data['obv_obv_signal_diff'].shift(1)
        data['obv_signal_call'] = data.apply(ca.obv_signal_call, axis=1) # GC
        data.pop('obv_obv_signal_diff')
        data.pop('obv_obv_signal_diff_shift')
        
        # VPT (Volume-price trend)
        data['vpt'] = ta.volume.volume_price_trend(close=data['Close'], volume=data['Volume'])
        data['vpt_signal'] = ta.trend.ema_indicator(close=data['vpt'], window=9 * 78) # GC
        data['vpt_vpt_signal_diff'] = data['vpt'] - data['vpt_signal'] # GC
        data['vpt_vpt_signal_diff_shift'] = data['vpt_vpt_signal_diff'].shift(1)
        data['vpt_signal_call'] = data.apply(ca.vpt_signal_call, axis=1) # GC
        data.pop('vpt_vpt_signal_diff')
        data.pop('vpt_vpt_signal_diff_shift')
        
        # VWAP (Volume Weighted Average Price)
        data['vwap'] = ta.volume.volume_weighted_average_price(high=data['High'], low=data['Low'], close=data['Close'], volume=data['Volume'], window=14 * 78)
        data['vwap_signal'] = ta.trend.ema_indicator(close=data['vwap'], window=9 * 78) # GC
        data['vwap_close_vwap_diff'] = data['Close'] - data['vwap'] # GC
        data['vwap_close_vwap_diff_shift'] = data['vwap_close_vwap_diff'].shift(1)
        data['vwap_vwap_signal_diff'] = data['vwap'] - data['vwap_signal']
        data['vwap_vwap_signal_diff_shift'] = data['vwap_vwap_signal_diff'].shift(1)
        data['vwap_call'] =  data.apply(ca.vwap_call, axis=1) # GC
        data['vwap_signal_call'] = data.apply(ca.vwap_signal_call, axis=1) # GC
        data.pop('vwap_close_vwap_diff')
        data.pop('vwap_close_vwap_diff_shift')
        data.pop('vwap_vwap_signal_diff')
        data.pop('vwap_vwap_signal_diff_shift')
        
        
        ''' 변동성 지표 (Volatility Indicator) '''
        # ATR (Average True Range)
        data['atr'] = ta.volatility.average_true_range(high=data['High'], low=data['Low'], close=data['Close'], window=14 * 78)
        
        # Bollinger Bands
        bollinger = ta.volatility.BollingerBands(close=data['Close'], window=20 * 78)
        data['bollinger_hband'] = bollinger.bollinger_hband()
        data['bollinger_mband'] = bollinger.bollinger_mavg()
        data['bollinger_lband'] = bollinger.bollinger_lband()
        data['bollinger_hband_indicator'] = bollinger.bollinger_hband_indicator() # GC
        data['bollinger_lband_indicator'] = bollinger.bollinger_lband_indicator() # GC
        data['bollinger_pband'] = bollinger.bollinger_pband()
        data['bollinger_wband'] = bollinger.bollinger_wband()
        data['bollinger_pband_signal'] = ta.trend.ema_indicator(close=data['bollinger_pband'], window=9 * 78) # GC
        data['bollinger_pband_signal_diff'] = data['bollinger_pband'] - data['bollinger_pband_signal']
        data['bollinger_pband_signal_diff_shift'] = data['bollinger_pband_signal_diff'].shift(1)
        data['bollinger_pband_shift'] = data['bollinger_pband'].shift(1)
        data['bollinger_pband_call'] = data.apply(ca.bollinger_pband_call, axis=1) # GC
        data['bollinger_pband_signal_call'] = data.apply(ca.bollinger_pband_signal_call, axis=1) # GC
        data.pop('bollinger_pband_shift')
        data.pop('bollinger_pband_signal_diff')
        data.pop('bollinger_pband_signal_diff_shift')
        
        # Donchian Channel
        # MEMO : Donchian은 다른 두 밴드 지표들과 다르게 새로운 고점을 찍었을때 쭉 오르지 않음. 따라서 신호선 사용 X
        donchian = ta.volatility.DonchianChannel(high=data['High'], low=data['Low'], close=data['Close'], window=20 * 78)
        data['donchian_channel_hband'] = donchian.donchian_channel_hband()
        data['donchian_channel_mband'] = donchian.donchian_channel_mband()
        data['donchian_channel_lband'] = donchian.donchian_channel_lband()
        data['donchian_channel_hband_indicator'] = data['Close'] >= data['donchian_channel_hband'] # GC
        data['donchian_channel_lband_indicator'] = data['Close'] <= data['donchian_channel_lband'] # GC
        data['donchian_channel_pband'] = donchian.donchian_channel_pband()
        data['donchian_channel_wband'] = donchian.donchian_channel_wband()
        # data['donchian_channel_pband_signal'] = ta.trend.ema_indicator(close=data['donchian_channel_pband'], window=9 * 78) # GC
        # data['donchian_channel_pband_signal_diff'] = data['donchian_channel_pband'] - data['donchian_channel_pband_signal']
        # data['donchian_channel_pband_signal_diff_shift'] = data['donchian_channel_pband_signal_diff'].shift(1)
        data['donchian_channel_pband_shift'] = data['donchian_channel_pband'].shift(1)
        data['donchian_channel_pband_call'] = data.apply(ca.donchian_channel_pband_call, axis=1) # GC
        # data['donchian_channel_pband_signal_call'] = data.apply(ca.donchian_channel_pband_signal_call, axis=1) # GC
        data.pop('donchian_channel_pband_shift')
        # data.pop('donchian_channel_pband_signal_diff')
        # data.pop('donchian_channel_pband_signal_diff_shift')
        
        # Keltner Channel
        keltner = ta.volatility.KeltnerChannel(high=data['High'], low=data['Low'], close=data['Close'], window=20 * 78, window_atr=10 * 78, original_version=False)
        data['keltner_channel_hband'] = keltner.keltner_channel_hband()
        data['keltner_channel_mband'] = keltner.keltner_channel_mband()
        data['keltner_channel_lband'] = keltner.keltner_channel_lband()
        data['keltner_channel_hband_indicator'] = keltner.keltner_channel_hband_indicator() # GC
        data['keltner_channel_lband_indicator'] = keltner.keltner_channel_lband_indicator() # GC
        data['keltner_channel_pband'] = keltner.keltner_channel_pband()
        data['keltner_channel_wband'] = keltner.keltner_channel_wband()
        data['keltner_channel_pband_signal'] = ta.trend.ema_indicator(close=data['keltner_channel_pband'], window=9 * 78) # GC
        data['keltner_channel_pband_signal_diff'] = data['keltner_channel_pband'] - data['keltner_channel_pband_signal']
        data['keltner_channel_pband_signal_diff_shift'] = data['keltner_channel_pband_signal_diff'].shift(1)
        data['keltner_channel_pband_shift'] = data['keltner_channel_pband'].shift(1)
        data['keltner_channel_pband_call'] = data.apply(ca.keltner_channel_pband_call, axis=1) # GC
        data['keltner_channel_pband_signal_call'] = data.apply(ca.keltner_channel_pband_signal_call, axis=1) # GC
        data.pop('keltner_channel_pband_shift')
        data.pop('keltner_channel_pband_signal_diff')
        data.pop('keltner_channel_pband_signal_diff_shift')
        
        # Ulcer Index
        data['ui'] = ta.volatility.ulcer_index(close=data['Close'], window=14 * 78)
        
        
        ''' 추세지표 (Trend Indicators) '''
        # ADX (Average Directional Movement Index)
        adx = ta.trend.ADXIndicator(high=data['High'], low=data['Low'], close=data['Close'], window=14 * 78)
        data['adx'] = adx.adx()
        data['adx_pos'] = adx.adx_pos()
        data['adx_neg'] = adx.adx_neg()
        data['adx_pos_neg_diff'] = adx.adx_pos() - adx.adx_neg() # GC
        data['adx_pos_neg_diff_shift'] = data['adx_pos_neg_diff'].shift(1)
        data['adx_pos_neg_call'] = data.apply(ca.adx_pos_neg_call, axis=1) # GC
        data.pop('adx_pos_neg_diff')
        data.pop('adx_pos_neg_diff_shift')
        
        # Aroon Indicator
        aroon = ta.trend.AroonIndicator(high=data['High'], low=data['Low'], window=25 * 78)
        data['aroon_up'] = aroon.aroon_up()
        data['aroon_down'] = aroon.aroon_down()
        data['aroon_up_down_diff'] = aroon.aroon_indicator() # GC
        data['aroon_up_down_diff_shift'] = data['aroon_up_down_diff'].shift(1)
        data['aroon_up_down_call'] = data.apply(ca.aroon_up_down_call, axis=1) # GC
        data.pop('aroon_up_down_diff')
        data.pop('aroon_up_down_diff_shift')
        
        # CCI (Commodity Channel Index)
        data['cci'] = ta.trend.cci(high=data['High'], low=data['Low'], close=data['Close'], window=20 * 78)
        data['cci_signal'] = ta.trend.ema_indicator(close=data['cci'], window=9 * 78) # GC
        data['cci_shift'] = data['cci'].shift(1)
        data['cci_cci_signal_diff'] = data['cci'] - data['cci_signal']
        data['cci_cci_signal_diff_shift'] = data['cci_cci_signal_diff'].shift(1)
        data['cci_call'] = data.apply(ca.cci_call, axis=1) # GC
        data['cci_signal_call'] = data.apply(ca.cci_signal_call, axis=1) # GC
        data.pop('cci_shift')
        data.pop('cci_cci_signal_diff')
        data.pop('cci_cci_signal_diff_shift')
        
        # DPO (Detrended Price Oscillator)
        data['dpo'] = ta.trend.dpo(close=data['Close'], window=20 * 78)
        data['dpo_shift'] = data['dpo'].shift(1)
        data['dpo_call'] = data.apply(ca.dpo_indicator, axis=1) # GC
        data.pop('dpo_shift')
        
        # 일목균형표 (Ichimoku Kinkō Hyō, Ichimoku)
        ichimoku = ta.trend.IchimokuIndicator(high=data['High'], low=data['Low'], window1=9 * 78, window2=26 * 78, window3=52 * 78)
        data['ichimoku_a'] = ichimoku.ichimoku_a()
        data['ichimoku_b'] = ichimoku.ichimoku_b()
        data['ichimoku_conversion_line'] = ichimoku.ichimoku_conversion_line()
        data['ichimoku_base_line'] = ichimoku.ichimoku_base_line()
        # data['ichimoku_close_position'] = data.apply(ca.ichimoku_close_position, axis=1) # GC
        data['ichimoku_conversion_base_diff'] = data['ichimoku_conversion_line'] - data['ichimoku_base_line'] # GC
        data['ichimoku_conversion_base_diff_shift'] = data['ichimoku_conversion_base_diff'].shift(1)
        data['ichimoku_close_a_diff'] = data['Close'] - data['ichimoku_a']
        data['ichimoku_close_b_diff'] = data['Close'] - data['ichimoku_b']
        data['ichimoku_close_a_diff_shift'] = data['ichimoku_close_a_diff'].shift(1)
        data['ichimoku_close_b_diff_shift'] = data['ichimoku_close_b_diff'].shift(1)
        data['ichimoku_conversion_base_call'] = data.apply(ca.ichimoku_conversion_base_call, axis=1) # GC
        data['ichimoku_a_b_call'] = data.apply(ca.ichimoku_a_b_call, axis=1) # GC
        data.pop('ichimoku_conversion_base_diff')
        data.pop('ichimoku_conversion_base_diff_shift')
        data.pop('ichimoku_close_a_diff')
        data.pop('ichimoku_close_a_diff_shift')
        data.pop('ichimoku_close_b_diff')
        data.pop('ichimoku_close_b_diff_shift')
        
        # KST Oscillator (KST Signal)
        kst = ta.trend.KSTIndicator(close=data['Close'], roc1=10 * 78, roc2=15 * 78, roc3= 20 * 78, roc4=30 * 78, window1=10 * 78, window2=10 * 78, window3=10 * 78, window4=15 * 78, nsig=9 * 78)
        data['kst'] = kst.kst()
        data['kst_signal'] = kst.kst_sig()
        data['kst_shift'] = data['kst'].shift(1)
        data['kst_kst_signal_diff'] = kst.kst_diff() # GC
        data['kst_kst_signal_diff_shift'] = data['kst_kst_signal_diff'].shift(1)
        data['kst_call'] = data.apply(ca.kst_call, axis=1) # GC
        data['kst_signal_call'] = data.apply(ca.kst_signal_call, axis=1) # GC
        data.pop('kst_shift')
        data.pop('kst_kst_signal_diff')
        data.pop('kst_kst_signal_diff_shift')
        
        # MACD (Moving Average Convergence Divergence)
        macd = ta.trend.MACD(data['Close'], window_slow=26 * 78, window_fast=12 * 78, window_sign=9 * 78)
        data['macd'] = macd.macd()
        data['macd_signal'] = macd.macd_signal()
        data['macd_shift'] = data['macd'].shift(1)
        data['macd_signal_diff'] = macd.macd_diff() # GC
        data['macd_signal_diff_shift'] = data['macd_signal_diff'].shift(1)
        data['macd_call'] = data.apply(ca.macd_call, axis=1) # GC
        data['macd_signal_call'] = data.apply(ca.macd_signal_call, axis=1) # GC
        data.pop('macd_shift')
        data.pop('macd_signal_diff')
        data.pop('macd_signal_diff_shift')
        
        # MI (Mass Index)
        data['mi'] = ta.trend.mass_index(high=data['High'], low=data['Low'], window_fast=9 * 78, window_slow=25 * 78) / 78
                
        # Parabolic SAR (Parabolic Stop and Reverse)
        # 5분 단위 데이터에서는 위아래로 진동을 함, smoothing을 해서 사용하여 해결
        psar = ta.trend.PSARIndicator(high=data['High'], low=data['Low'], close=data['Close'])
        data['psar'] = psar.psar()
        data['psar_up'] = psar.psar_up()
        data['psar_down'] = psar.psar_down()
        data['psar_up_indicator'] = psar.psar_up_indicator() # GC
        data['psar_down_indicator'] = psar.psar_down_indicator() # GC
        data['psar_smoothing'] = ta.trend.ema_indicator(close=ta.trend.ema_indicator(close=data['psar'], window=1 * 78), window=1 * 78)
        data['psar_close_psar_diff'] = data['Close'] - data['psar_smoothing'] # GC
        data['psar_close_psar_diff_shift'] = data['psar_close_psar_diff'].shift(1)
        data['psar_call'] = data.apply(ca.psar_call, axis=1) # GC
        data.pop('psar_up')
        data.pop('psar_down')
        data.pop('psar_up_indicator')
        data.pop('psar_down_indicator')
        data.pop('psar_close_psar_diff')
        data.pop('psar_close_psar_diff_shift')
        
        # STC (Schaff Trend Cycle)
        data['stc'] = ta.trend.stc(close=data['Close'], window_slow=50 * 78, window_fast=23 * 78, cycle=10 * 78, smooth1=3 * 78, smooth2=3 * 78)
        data['stc_shift'] = data['stc'].shift(1)
        data['stc_call'] = data.apply(ca.stc_indicator, axis=1) # GC
        data.pop('stc_shift')
        
        # TRIX (Triple Exponential Average)
        data['trix'] = ta.trend.trix(close=data['Close'], window=15 * 78)
        data['trix_signal'] = ta.trend.ema_indicator(close=data['trix'], window=9 * 78) # GC
        data['trix_shift'] = data['trix'].shift(1)
        data['trix_trix_signal_diff'] = data['trix'] - data['trix_signal'] # GC
        data['trix_trix_signal_diff_shift'] = data['trix_trix_signal_diff'].shift(1)
        data['trix_call'] = data.apply(ca.trix_call, axis=1)
        data['trix_signal_call'] = data.apply(ca.trix_signal_call, axis=1) # GC
        data.pop('trix_shift')
        data.pop('trix_trix_signal_diff')
        data.pop('trix_trix_signal_diff_shift')
        
        # VI (Vortex Indicator)
        vi = ta.trend.VortexIndicator(high=data['High'], low=data['Low'], close=data['Close'], window=14 * 78)
        data['vi_pos'] = vi.vortex_indicator_pos()
        data['vi_neg'] = vi.vortex_indicator_neg()
        data['vi_pos_neg_diff'] = vi.vortex_indicator_diff() # GC
        data['vi_pos_neg_diff_shift'] = data['vi_pos_neg_diff'].shift(1)
        data['vi_pos_neg_call'] = data.apply(ca.vi_call, axis=1) # GC
        data.pop('vi_pos_neg_diff')
        data.pop('vi_pos_neg_diff_shift')
                
    
        ''' 기타 지표 (Others Indicator)'''
        # 없음
        
        
        ''' 임의 생성 지표 '''
        # CMO (Chande Momentum Oscillator)
        # MEMO : RSI처럼 5분 단위 데이터로 진입했을 때 값이 쪼그라듬, 0선 교차는 너무 느리게 반응 하는 것 같고, -50, 50선 교차는 값이 쪼그라들어 사용할 수 없을 듯. 신호선만 사용
        data['cmo'] = ca.cmo(close=data['Close'], window=14 * 78)
        data['cmo_signal'] = ta.trend.ema_indicator(close=data['cmo'], window=9 * 78)
        # data['cmo_shift'] = data['cmo'].shift(1)
        data['cmo_cmo_signal_diff'] = data['cmo'] - data['cmo_signal']
        data['cmo_cmo_signal_diff_shift'] = data['cmo_cmo_signal_diff'].shift(1)
        # data['cmo_call'] = data.apply(ca.cmo_call, axis=1) # GC
        data['cmo_signal_call'] = data.apply(ca.cmo_signal_call, axis=1) # GC
        # data.pop('cmo_shift')
        data.pop('cmo_cmo_signal_diff')
        data.pop('cmo_cmo_signal_diff_shift')
        
        # Pivot Point
        # MEMO : 1번만 쓰자
        data['pp1'] = (data['High'] + data['Low'] + data['Close']) / 3
        data['pp1_close_pp1_diff'] = data['Close'] - data['pp1']
        data['pp1_close_pp1_diff_shift'] = data['pp1_close_pp1_diff'].shift(1) 
        # data['pp2'] = (data['Open'] + data['High'] + data['Low'] + data['Close']) / 4
        # data['pp2_close_pp2_diff'] = data['Close'] - data['pp2']
        # data['pp2_close_pp2_diff_shift'] = data['pp2_close_pp2_diff'].shift(1) 
        data['pp1_call'] = data.apply(ca.pp1_call, axis=1) # GC
        # data['pp2_call'] = data.apply(ca.pp2_call, axis=1) # GC
        data.pop('pp1_close_pp1_diff')
        data.pop('pp1_close_pp1_diff_shift')
        
        # # 캔들스틱 차트 (Candlestick chart)
        # data['doji'] = data.apply(ca.is_doji, axis=1)  # 도지 (Doji) 패턴
        # data['hammer'] = data.apply(ca.is_hammer, axis=1)  # 해머 (Hammer)
        # data['inverted_hammer'] = data.apply(ca.is_inverted_hammer, axis=1)  # 역해머 (Inverted Hammer)
        # data['morning_star'] = [ca.is_morning_star(data, i) for i in range(len(data))]  # 유성형 (Morning Star)
        # data['triple_top'] = [ca.is_triple_top(data, i) for i in range(len(data))]  # 삼중천장 (Triple Top)
        # data['triple_tottom'] = [ca.is_triple_bottom(data, i) for i in range(len(data))]  # 삼중바닥 (Triple Bottom)
        # data['red_marubozu'] = data.apply(ca.is_red_marubozu, axis=1)  # 붉은색 마루보즈 (Red Marubozu)
        # data['green_marubozu'] = data.apply(ca.is_green_marubozu, axis=1)  # 녹색 마루보즈 (Green Marubozu)
        # data['three_black_crows'] = [ca.is_three_black_crows(data, i) for i in range(len(data))]  # 흑삼병 (Three Black Crows)
        # data['three_white_soldiers'] = [ca.is_three_white_soldiers(data, i) for i in range(len(data))]  # 적삼병 (Three White Soldiers)
        
        
        # 데이터 정리
        data = pd.concat([data[['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']], data[['atr', 'bollinger_wband', 'donchian_channel_wband', 'keltner_channel_wband', 'ui', 'adx', 'adx_pos', 'adx_neg', 'aroon_up', 'aroon_down']], data.filter(like='_call')], axis=1)
        data.columns = pd.MultiIndex.from_product([data.columns, [ticker]]) # 컬럼 레벨 다시 ticker로 추가
        data.columns.names = ['', '']
        
        
        # 데이터 저장
        data.to_csv(f'Data/Call/{ticker}_call_5min.csv', index=True)
        print(f'{ticker} data saved')
```
